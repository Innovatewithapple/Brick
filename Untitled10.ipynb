```python
import tensorflow as tf
from tensorflow.keras import layers
```

```python
vocab_size = 200
max_length = 10000
embed_dim = 64
ff_dim = 128
num_head = 4
```

```python
def Transformer_Block(inputs):
  #Brick2: multihead QKV
  attentionOuptut = layers.MultiHeadAttention(num_heads=num_head,key_dim=embed_dim)
  attentionOutput = attentionOuptut(query=inputs,key=inputs,value=inputs)
  attentionOutput = layers.Dropout(0.1)(attentionOutput)

  #Brick4: Part one Glue
  x1 = layers.LayerNormalization(epsilon=1e-6)(inputs + attentionOutput)


  #Brick3: MLP
  x2 = layers.Dense(ff_dim,activation='relu')(x1)
  x2 = layers.Dense(embed_dim)(x2)
  output = layers.Dropout(0.1)(x2)

  return layers.LayerNormalization(epsilon=1e-6)(x1 + x2)
```

```python
#Brick1: token embedding
input = layers.Input(shape=(max_length,))

token_embed = layers.Embedding(vocab_size,embed_dim)(input)

#Position embedding
position = tf.range(start=0,limit=max_length,delta=1)
postion_embed = layers.Embedding(max_length,embed_dim)(position)

merge = token_embed + postion_embed

x = Transformer_Block(merge)

x = layers.GlobalAveragePooling1D()(x)
x = layers.Dropout(0.1)(x)
x = layers.Dense(20,activation='relu')(x)
final = layers.Dense(2,activation='softmax')(x)

model = tf.keras.Model(inputs = input,outputs= final)
model.compile('adam',loss='sparse_category_classentropy',metrics=['accuracy'])
```
