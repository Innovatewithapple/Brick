```python
import tensorflow as tf
from tensorflow.keras import layers
```

```python
#The setup (Hyperparameters)

vocab_size = 10000 # the number of total words (or number of patches from image) to consider in the project
max_length = 200 # maximum number of words in a particular sentence or the sequence
embed_dim = 32 # number of features we want from particular word
num_heads = 2 #to divide the features equally to get more information
ff_dim = 128 # Basically we expand the features one last time to get more information and here we use 4x value of our feature, and then after extraction, convert back to again our original embed_dim
```

```python
#Transformer block

def TransformerBlock(inputs):
  #Brick2: Attention layer Q,K,V
  attention_output = layers.MultiHeadAttention(num_heads=num_heads,key_dim=embed_dim)
  attention_output = attention_output(query=inputs,key=inputs,value=inputs)
  attention_output = layers.Dropout(0.1)(attention_output)


  #Brick4: Glue, (Part:A) first rule add token embedding-position embedding with attention output to get normalizaiton (means small or large feature values to convert around 0 to 1)
  x1 = layers.LayerNormalization(epsilon=1e-6)(inputs + attention_output)

  #Brick3: MLP, Here we expand our features more to get information and convert back to its original shape
  x2 = layers.Dense(ff_dim,activation='relu')(x1)
  x2 = layers.Dense(embed_dim)(x2)
  x2 = layers.Dropout(0.1)(x2)

  return layers.LayerNormalization(epsilon=1e-6)(x1 + x2)
```

The final pipeline

```python
inputs = layers.Input(shape=(max_length,))

#Brick1: Token Embedding and Positional Embedding. And then we add them so that each word has now the position to determine in the sentence
token_embedding = layers.Embedding(vocab_size,embed_dim)(inputs) # here basically we tell that how many total words are to get features. And input is basically the length of the input can pass

positional_input = tf.range(start=0,limit=max_length,delta=1)
positional_embedding = layers.Embedding(max_length,embed_dim)(positional_input) # here basically we create positioning with same length of the sentence
x = token_embedding + positional_embedding

#Now we call Transformer block. And we can call it as many time we want.
x = TransformerBlock(x)

#The Exit: Classifier Head
x = layers.GlobalAveragePooling1D()(x)
x = layers.Dropout(0.1)(x)
x = layers.Dense(20,activation='relu')(x)
outputs = layers.Dense(2,activation='softmax')(x)

model = tf.keras.Model(inputs=inputs,outputs=outputs)
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```
